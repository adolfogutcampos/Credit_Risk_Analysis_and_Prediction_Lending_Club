## 5.1.- Loan ammount (Monto del prestamo)

Crear bins para loan_amnt_range

```python
bins = list(range(0, 45000, 5000))
labels = [f'{i}-{i+5000}' for i in range(0, 40000, 5000)]
```

## 5.2.- Annual Income (Ingresos Anuales)

```python
annual_income_bin = data['annual_inc'].value_counts(bins=30).sort_values()
annual_income_bin.tail(20)
```

```python
bins_annual_income = [0, 0.1, 20000, 40000, 60000, 80000, 100000, 120000, 140000, 160000, 180000, 200000, float('inf')]
labels_annual_income = [
    '0', '0.1-20000', '20000-40000', '40000-60000', 
    '60000-80000', '80000-100000', '100000-120000', 
    '120000-140000', '140000-160000', '160000-180000', 
    '180000-200000', '200000+'
]
```

```python
data['annual_inc_range'] = pd.cut(data['annual_inc'], bins=bins_annual_income, labels=labels_annual_income, right=False, include_lowest=True)
```

```python
annual_inc_range_counts = data['annual_inc_range'].value_counts().sort_index()
```

```python
total_count = annual_inc_range_counts.sum()
percentage_labels = (annual_inc_range_counts / total_count * 100).round(2).astype(str) + '%'
```

```python
fig = go.Figure()

fig.add_trace(go.Bar(
    x=annual_inc_range_counts.index.astype(str),
    y=annual_inc_range_counts.values,
    marker=dict(color=annual_inc_range_counts.values, colorscale='Viridis'),
    text=[f'{count}' for count in annual_inc_range_counts.values],
    textposition='inside',
    insidetextanchor='middle'
))

for i, (x, y, percentage) in enumerate(zip(annual_inc_range_counts.index.astype(str), annual_inc_range_counts.values, percentage_labels)):
    fig.add_annotation(
        x=x,
        y=y,
        text=percentage,
        showarrow=False,
        yshift=10,
        font=dict(size=12)
    )

fig.update_layout(
    title='Distribution of Annual Income',
    xaxis_title='Annual Income Range (USD)',
    yaxis_title='Frequency',
    title_font_size=16,
    width=1200,
    height=600,
    bargap=0.2  
)

fig.show()
```

```python
data['FICO_average'] = (data['fico_range_high'] + data['fico_range_low'])/2
data['FICO_average'].head()
```

```python
bins_fico = list(range(500, 925, 25))
labels_fico = [f'{i}-{i+25}' for i in range(500, 900, 25)] + ['900+']
```

```python
total_count = fico_range_counts.sum()
percentage_labels = (fico_range_counts / total_count * 100).round(2).astype(str) + '%'
```
